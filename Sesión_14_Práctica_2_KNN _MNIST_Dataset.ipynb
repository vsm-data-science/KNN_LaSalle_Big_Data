```python
mnist_train = pd.read_csv('./mnist_train.csv', encoding='utf-8', sep=',')
mnist_train.head(9)
```

```python
mnist_test = pd.read_csv('./mnist_test.csv', encoding='utf-8', sep=',')
mnist_test.head(9)
```

## **Machine Learning** | Entrenamos el modelo

Vamos a dividir el conjunto de datos en dos partes: conjunto de características y etiqueta de destino

```python
feature_train = mnist_train.loc[:,mnist_train.columns !='label']
target_train = mnist_train.label
```

```python
feature_test = mnist_test.loc[:,mnist_test.columns !='label']
target_test = mnist_test.label
```

**K-Nearest Neighbors Classification**

```python
# Create a KNN classifier object
knn = KNeighborsClassifier(n_neighbors=2)
 
# Train the model using the training dataset
knn.fit(feature_train,target_train)
 
# Predict the target variable for test dataset
predictions_knn = knn.predict(feature_test)
```

```python
predictions_knn
```

```python
# Import metrics module for performance evaluation
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
 
# Calculate model accuracy
print("Accuracy:",accuracy_score(target_test, predictions_knn))
```

```python
# Calculate model precision
print("Precision:",precision_score(target_test, predictions_knn, average='micro'))
# Calculate model recall
print("Sensitivity/Recall:",recall_score(target_test, predictions_knn, average='micro'))
# Calculate model f1 score
print("F1-Score:",f1_score(target_test, predictions_knn, average='micro'))
```
